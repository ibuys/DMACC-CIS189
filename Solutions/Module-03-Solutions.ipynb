**1. Basic if-elif Statement Assignment**

Write a program that asks for the user to sign up for Programmer's Toolkit Monthly Subscription Box. They must select the level of membership they want. Each month is something different, t-shirts, stickers, figurines, even programming books!

The levels are the following:

- Platinum
- Gold
- Silver
- Bronze
- Free Trial


Write an if statement that prints the cost for each of the level. Platinum is $60, each level below is 10 dollars cheaper, and the free trial is free.

```python

level = input("Enter membership level: ").strip().title()

if level == "Platinum":
    cost = 60

elif level == "Gold":
    cost = 50

elif level == "Silver":
    cost = 40

elif level == "Bronze":
    cost = 30

elif level == "Free Trial":
    cost = 0

print(f"Membership level: {level}, cost = ${cost}")
```

Write code to perform the following:

- Set MAX = 10, MIN = 0
- Set a variable y to a value 
- Test whether y is above MAX
- Test whether y is below MIN
- Set a variable x to a value
- Test whether it is between the MIN and MAX but does not equal MAX nor MIN
- Test whether it is within MIN up to MAX but does not equal MAX  
- Test whether it is above MIN up to and including MAX

```python

MIN = 0
MAX = 10

y = 7

# Test whether y is above MAX.
if y > MAX:
    print(f"{y} greater than {MAX}.")
else:
    print(f"{y} less than or equal to {MAX}.")

# Test whether y is below MIN.
if y < MIN:
    print(f"{y} less than {MIN}.")
else:
    print(f"{y} greater than or equal to {MIN}.")


# Set a variable x to a value.
x = 3

# Test whether x is between the MIN and MAX but does not equal MAX nor MIN.
if MIN < x < MAX:
    print(f"{x} between ({MIN}, {MAX}).")

# Test whether it is within MIN up to MAX but does not equal MAX.
if MIN <= x < MAX:
    print(f"{x} between [{MIN}, {MAX}).")

# Test whether it is above MIN up to and including MAX.
if MIN < x <= MAX:
    print(f"{x} between ({MIN}, {MAX}].")
    
```


**Nested if statement Assignment**

An online shopping retailer offer two types of discounts, cash-off coupons and percent discount coupons. You must add tax after applying coupons. If you have cash-off coupons, those must be applied first, then apply the percent discount coupons on the pre-tax amount.

The following is the order of calculations: 

- The first is cash-off coupons. You may apply one $5 or $10 cash off coupon per order.
- The second is percent discount coupons for 10%, 15%, or 20% off.
- The third is tax, 6% of the cost after cash-off and percent discount coupons.
- The fourth is shipping cost (applied to pre-tax amount from step 2):

    - up to $10 dollars, shipping is $5.95
    - $10 and up to $30 dollars, shipping is $7.95
    - $30 and up to $50 dollars, shipping is $11.95
    - Shipping is free for $50 and over
    
Write a script that prompts the user for the the amount of the original purchase price, the cash coupon and the percent coupon. Then calculate and return the total order price for the item.

For instance, you have $5 off and 30% coupons and the item is $15.99, you will first take:

- $15.99 - $5.00 = $10.99.
- 30% off $10.99 to get $7.69
- Calculate tax at 6%  (.06 * 7.69 = 0.46)
- Add shipping cost to pre-tax price (which is $7.69 in this example):  $5.95 + $7.69 = $13.64.
- Total cost = 13.64 + .46 = 14.10.

Add shipping cost to pre-tax price (which is $7.69 in this example):  $5.95 + $7.69 = $13.64
Do not round intermediate values. Your final result should be rounded to 2 decimal places. In order to receive full credit, your code must return the correct final price to 2 decimal places. 

Do not forget to use variables and constants where appropriate.



```python

orig_price = float(input("Enter original purchase price: ").strip())
first_coupon = float(input("Enter cash-off coupon (5|10): ").strip())
second_coupon = float(input("Enter percent-off coupon as decimal (.10|.15|.20): ").strip())

# Subtract first coupon from orig_price. We can ignore it for now, but 
# we might want to handle the case in which first_coupon exceeds orig_price.
discounted_price = orig_price - first_coupon

# Apply percent-off coupon.
discounted_price = (1 - second_coupon) * discounted_price 

# Tax amount = 6% of the cost after cash-off and percent discount coupons.
tax_amount = .06 * discounted_price

# Determine shipping cost.
if 0 < discounted_price <= 10:
    shipping_amount = 5.95

elif 10 < discounted_price <= 30:
    shipping_amount = 7.95

elif 30 < discounted_price <= 50:
    shipping_amount = 11.95

else:
    shipping_amount = 0.00

final_cost = discounted_price + tax_amount + shipping_amount


print(f"discounted_price: {discounted_price:,.2f}")
print(f"tax_amount: {tax_amount:,.2f}")
print(f"shipping_amount: {shipping_amount}")
print(f"Final cost: {final_cost:,.2f}")
```
