Complete the `average_scores` functions using arbitrary argument lists and keyword arguments and returning a string in the following format (for example, you will have your own):

> 'Result: name = M gpa = 3.2 course = Python with current average 30.0

The keywords will include student name, course name, current gpa. The example above used name, gpa, and course. You can select your own.

````python

def average_score(*args, **kwargs):
    """
    Compute average scores for a variable number of test scores. Keyword arguments
    may include name, course and gpa.
    """
    avg = sum(args) / len(args)

    # Unpack keyword arguments. Use default values if keyword argument not present. 
    # course = kwargs.get("course", "CIS189")
    # name = kwargs.get("name", "John Smith")
    # gpa = kwargs.get("gpa", 4.0)
    gpa = kwargs["gpa"]
    name = kwargs["name"]
    course = kwargs["course"]

    msg = f"name={name}, gpa={gpa}, course={course}, average={avg:.2f}"

    return msg


msg1 = average_score(71, 81, 91, 99, 45, course="puppy", name="JDT", gpa=3.98)
msg2 = average_score(71, 81, 96)
msg3 = average_score(79, name="Tommy")

print(msg1)
print(msg2)
print(msg3)
````

````python
"""
CIS189 Module 7 File I/O Assignment.
"""

# ------------------------------------------------------------------------------
# Part 1.: Complete write_to_file function (void).
# ------------------------------------------------------------------------------

def write_to_file(scores, name="some-student-name"):
    """
     Parameters
     ----------
     scores: list
         List of scores.
     name: str
         Student name.
    """
    fname = f"{name}.txt"
    with open(fname, "w") as f:
        for score in scores:
            f.write(str(score) + "\n")




# ------------------------------------------------------------------------------
# Part 2: Complete get_student_info function (void).
# ------------------------------------------------------------------------------
     
def get_student_info():
    """
    ##### ADD DOCSTRING HERE #####
    """
    # Create empty list to hold scores.
    scores = []

    # Prompt for student name.
    name = input("Enter student name: ")

    # Prompt for number of scores to enter.
    nbr_scores = int(input("How many scores do you wish to enter? "))

    # Iterate up to nbr_scores, prompting user to enter a score each time.
    for i in range(nbr_scores):

        # Prompt for score.
        score = float(input(f"Enter score {i + 1} for {name}: "))

        # Append score to scores list initialized above.
        scores.append(score)

    # Pass student scores and name to write_to_file function.
    write_to_file(scores, name=name)




# ------------------------------------------------------------------------------
# Part 3: Complete read_from_file function (fruitful).
# ------------------------------------------------------------------------------

def read_from_file(filename):
    """
    Load scores from filename and return string containing filename, 
    min score, max score and average score.

    Parameters
    ----------
    filename: str
        Name of file with scores, assumed to have .txt extension.

    Returns
    -------
    str
        String containing filename, min score, max score and average score.
    """
    # Open filename for reading. Load scores into list performing necessary 
    # type conversions.
    with open(filename, "r") as f:
        scores = f.readlines()
        scores = [float(ii.strip()) for ii in scores]

    # Identify min, max and average score from list of scores.
    min_score = min(scores)
    max_score = max(scores)
    avg_score = sum(scores) / len(scores)

    # Return string formatted as:
    # {filename}: min={min_score}, max={max_score} avg={avg_score}
    msg = f"{filename}: min={min_score}, max={max_score}, avg={avg_score}"
    return msg




get_student_info() # john.txt
get_student_info() # paul.txt
get_student_info() # george.txt


# Call read_from_file for created text files and print the results.
print(read_from_file("john.txt"))
print(read_from_file("paul.txt"))
print(read_from_file("george.txt"))
````
