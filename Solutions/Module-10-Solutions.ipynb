**1.b** Complete the `display` method (fruitful) within the Employee class, which returns a string that displays information for the employee in the following format:

- LAST_NAME, FIRST_NAME: <last name, first name here>
- PHONE_NUMBER: <phone number here>
- ADDRESS: <address here>
- START DATE: <start date here>
- SALARY: <salary here>

  

```python


class Employee:
    """
    Employee class.
    """
    def __init__(self, last_name, first_name, address, phone_number, salary, start_date):
        self.last_name = last_name
        self.first_name = first_name
        self.address = address
        self.phone_number = phone_number
        self.salary = salary
        self.start_date = start_date


    def display(self):
        """
        Display employee information as:

            LAST_NAME, FIRST_NAME: <last name, first name here>
            PHONE_NUMBER: <phone number here>
            ADDRESS: <address here>
            START DATE: <start date here>
            SALARY: <salary here>
        """
        name_str = f"- LAST_NAME, FIRST_NAME: {self.last_name}, {self.first_name}"
        phone_str = f"- PHONE_NUMBER: {self.phone_number}"
        addr_str = f"- ADDRESS: {self.address}"
        date_str = f"- START_DATE: {self.start_date}"
        salary_str = f"- SALARY: {self.salary}"
        emp_str = f"{name_str}\n{phone_str}\n{addr_str}\n{date_str}\n{salary_str}"
        return emp_str

```

**2. Invoice Class Assignment**

```python

class Invoice:

    def __init__(self, invoice_id, customer_id):
        self.invoice_id = invoice_id
        self.customer_id = customer_id

        # Initialize empty dictionary.
        self.ditems = {}


    def add_item(self, item, price) -> None:
        """
        Add item, price to self.ditems as self.ditems[item] = price.
        
        Parameters
        ----------
        item: str
            Item description.

        price: float
            Price of item.
        """
        self.ditems[item] = price

    
    def print_invoice(self):
        """
        Print items in invoice in ascending order. For example:

            CUSTOMER: 98676867
            INVOICE: A554GFT-117
            item_name_1: 199.99
            item_name_2: 249.50
            item_name_3: 599.67
        """
        cust_str = f"CUSTOMER: {self.customer_id}"
        inv_str = f"INVOICE: {self.invoice_id}"
        pairs = sorted(self.ditems.items(), key=lambda v: v[1])

        print(cust_str)
        print(inv_str)

        for item, price in pairs:
            print(f"{item}: ${price:.2f}")


    def __repr__(self):
        return f"Invoice({self.invoice_id}, {self.customer_id})" 
    
    
    def __str__(self):
        return f"invoice-id: {self.invoice_id}, customer-id: {self.customer_id})" 



def main():

    invoices_list = [
        {"invoice_id": "A554GFT-117", "customer_id": "98676867", "items": [("ipad", 499.99), ("mouse", 29.97), ("monitor", 229.08)]},
        {"invoice_id": "J244GFT-283", "customer_id": "77594872", "items": [("router", 267.78), ("hdmi-cable", 24.77)]},
        {"invoice_id": "R943RXC-476", "customer_id": "54092341", "items": [("laptop", 1399.54), ("warranty", 199.94), ("snickers", 2.49)]},
        {"invoice_id": "J712WQR-888", "customer_id": "43908134", "items": [("amplifier", 749.00), ("speaker-cable", 42.75)]}
        ]

    for dinv in invoices_list:

        invoice_id = dinv["invoice_id"]
        customer_id = dinv["customer_id"]
        pairs = dinv["items"]  # list of 2-tuples.

        # Create Invoice object.
        inv = Invoice(invoice_id, customer_id)

        # Add items to instance dictionary.
        for item, price in pairs:
            inv.add_item(item, price)

        # Call print_invoice.
        inv.print_invoice()

        print("\n")


main()
```

**3. Unit Tests for a Class Assignment**

```python

import unittest


class Student:
    """Student class"""
    def __init__(self, lname, fname, major, gpa=0.0):

        if not (lname or isinstance(lname, str)):
            raise ValueError 
        if not (fname or isinstance(fname, str)):
            raise ValueError
        if not (major or isinstance(major, str)):
            raise ValueError
        if not isinstance(gpa, (int, float)) or not 0 <= gpa <=4.0:
            raise ValueError

        self.last_name = lname
        self.first_name = fname
        self.major = major
        self.gpa = gpa

    def __str__(self):
        return self.last_name + ", " + self.first_name + " has major " + self.major + " with gpa: " + str(self.gpa)
    


class StudentTestCase(unittest.TestCase):

    def setUp(self):
        self.student = Student("Triveri", "James", "Physics", 3.89)

    def tearDown(self):
        del self.student

    def test_object_created_required_attributes(self):
        self.assertEqual(self.student.last_name, "Triveri")
        self.assertEqual(self.student.first_name, "James")
        self.assertEqual(self.student.major, "Physics")
        
    def test_object_created_all_attributes(self):
        self.assertEqual(self.student.last_name, "Triveri")
        self.assertEqual(self.student.first_name, "James")
        self.assertEqual(self.student.major, "Physics")
        self.assertEqual(self.student.gpa, 3.89)

    def test_student_str(self):
        self.assertEqual("Triveri, James has major Physics with gpa: 3.89", str(self.student))

    def test_object_not_created_error_last_name(self):
        with self.assertRaises(ValueError):
            s = Student(None, "James", "Physics")

    def test_object_not_created_error_first_name(self):
        with self.assertRaises(ValueError):
            s = Student("Triveri", None, "Physics")

    def test_object_not_created_error_major(self):
        with self.assertRaises(ValueError):
            s = Student("Triveri", "James", None)

    def test_object_not_created_error_gpa(self):
        with self.assertRaises(ValueError):
            s = Student("Triveri", "James", "Physics", "3.75")
```
