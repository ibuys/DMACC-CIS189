**1. Composition Assignment**

```python
"""
CIS189 Module 11 Composition assignment script.
"""
import random
from datetime import datetime



# Do not modify.
class Memory:
    """
    Class representation of computer memory.
    """
    def __init__(self):

        self.total_ram_gb = 16

    def available(self):
        in_use = random.uniform(0, self.total_ram_gb)
        avail = self.total_ram_gb - in_use 
        return avail


# Do not modify.
class DiskSpace:
    """
    Class representation of computer hard disk.
    """
    def __init__(self):
        self.total_capacity_gb = 512

    def available(self):
        in_use = random.uniform(0, self.total_capacity_gb)
        avail = self.total_capacity_gb - in_use 
        return avail


class Computer:
    """
    Computer class. 
    
    Attributes:
        name: str
            Passed in as an argument to the constructor. 

    Methods: 
        get_info:
            Returns a string of the form:

            <timestamp>: computer name: <name>, avail. memory: <available memory>, available storage: <available storage>
    
    """
    def __init__(self, name):
        self.name = name
        self.ram = Memory()
        self.storage = DiskSpace()

    def get_info(self):
        ts = datetime.now().strftime("%c")
        avail_ram = self.ram.available()
        avail_stor = self.storage.available()
        print(f"{ts}: computer name: {self.name}, avail. memory: {avail_ram:,.2f}GB, avail. storage: {avail_stor:,.2f}GB")


def main():


    machine_names = ["RQS445", "MIKES-MACHINE", "Leet-315"]

    # For each of the machine_names, create an instance of the Computer
    # class, then call `get_info` for each. 
    for machine in machine_names:
        c = Computer(machine)
        c.get_info()


main()
```

**2. Derived Class Assignment**

```python
"""
CIS189 Module 11 derived class assignment solution.
"""


class Person:
    def __init__(self, last_name, first_name):
        self.last_name = last_name
        self.first_name = first_name
    
    def __repr__(self):
        return(f"Person({self.last_name}, {self.first_name})")


class Student(Person):
    """
    Student class.

    Accepts:
        - Person instance
        - major
        - gpa

    Implements:
        - change_major: Updates the student's major (no return value).
        - update_gpa: Updates the student's gpa (no return value).
        - display: Prints the student's first name, last name, major and gpa. 
    """
    def __init__(self, last_name, first_name, major, gpa):
        super().__init__(last_name, first_name)
        self.major = major
        self.gpa = gpa

    def change_major(self, new_major):
        self.major = new_major

    def update_gpa(self, new_gpa):
        self.gpa = new_gpa

    def display(self):
        print(f"{self.last_name}, {self.first_name}: major={self.major}, gpa={self.gpa}")




def main():

    new_student = {"last_name": "Owens", "first_name": "Bonnie", "major": "Mathematics", "gpa": 4.0}

    # 1. Create new Person and Student instances using the information from new_student.
    last_name = new_student["last_name"]
    first_name = new_student["first_name"]
    major = new_student["major"]
    gpa = new_student["gpa"]

    s = Student(last_name, first_name, major, gpa)

    # 2. Call your Student's display method to print the current first name, last name, major and gpa. 
    s.display()
    
    # 3. Update the student's major and gpa using the following information. Once updated, 
    #    call your Student's display method again to verify that major and gpa have been 
    #    updated. 
    new_major = "Anthropology"
    new_gpa = 3.75

    s.change_major(new_major)
    s.update_gpa(new_gpa)
    s.display()


main()
```
